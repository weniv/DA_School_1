#### 해변 데이터 중 검색량의 이상치 판별

혼잡도로 표현

```python
# q1, q2, q3 * 1.5 -> 초과: 이상치 -> 검색량이 너무 많다 -> 혼잡도가 높다.

plt.boxplot(nature['검색건수'])
plt.show()
```

```python
# 상위 경계값 -> 혼잡도가 높다
# q3 - q1
IQR = 53093 - 6737
upper_IQR = (IQR * 1.5) + 53093
upper_IQR
```

```python
import numpy as np
```

```python
# 상위 경계값을 초과하는 값들을 지워준다.
# 이상치이 때문에 인기가 많고 혼잡할 것이다.
nature['혼잡도'] = np.where(nature['검색건수'] > upper_IQR, '혼잡', '보통')
nature
# 놀 수 있는 바다에 대한 순위 정보
```

### 양성, 음성 검사자 구분하기

```python
# 데이터 타입들 중 컬럼들이 서로 상관성이 있거나 반대 성향(경향)이 있으면 서로 비교 -> 상관관계 , 상관계수
pos = kr['확진자']
neg = kr['결과 음성']
```

### 사망율(치명율) 구하기

```python
# 두 데이터(컬럼) 합쳐서 하나의 결과를 도출 -> 컬럼 선택
# 확진자 수 중 사망 -> 사망율, 치명율
# 환진자 수: pos
pos
death = kr['사망']
# death/pos * 100 -> 비율
```

```python
d_p = round((death / pos) * 100, 2)
d_p.head()
```

### 완치율 구하기

```python
# 완치율
# 확진자 수
# 격리해제 수
pos
rec = kr['격리해제']

# rec/pos * 100 -> 비율
```

```python
r_p = round((rec / pos) * 100, 2)
r_p.head()
```

```python
# 치명율, 완치율 -> 그래프로 그려봤을 때 완치율이 눈에 띄게(유의미하게) 증가

plt.plot(d_p)
plt.plot(r_p)
plt.show()
```
