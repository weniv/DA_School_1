```python
# 1반 학생들의 국어(수학, 영어) 성적 평균
# 과목별로 최대 최소값
# 순서대로 -> 사분위값 -> 상위 몇 %
```

```python
# 1반 학생들의 국어(수학, 영어) 성적 평균
m = out_df.mean()

# bar 그래프 bar(x, y)
```

```python
m.index
```

```python
# plt.bar(m.index, m)
plt.bar(out_df.columns, m)
plt.show()
```

```python
out_df.describe()
```

```python
# 과목별로 최대 최소값
print(out_df.max())
print(out_df.min())
```

```python
# 순서대로 -> 사분위값 -> 상위 몇 %
# q1(25%), q2(50%), q3(75%)

q1 = out_df.quantile(0.25)
q1
```

```python
q2 = out_df.quantile(0.5) 
q3 = out_df.quantile(0.75)
print(q2)
print(q3)
```

```python
# IQR: q3 - q1 -> 50%에 해당되는 데이터
IQR = q3 - q1
# [-(1.5 * IQR) < 데이터 범위 < +(1.5 * IQR)] 일반적인 데이터 허용 범위: 이 범위를 벗어나면 이상치(Outlier)
IQR
```

```python
# 국어 점수 이상치 판별 함수
def kor_outlier(df):
    kor_score = df['국어']
    if kor_score > q3['국어'] + 1.5 * IQR['국어'] or kor_score < q1['국어'] - 1.5 * IQR['국어']:
        return True
    return False

out_df['국어_이상치'] = out_df.apply(kor_outlier, axis=1)
out_df
```

### 이상치 확인하기

- 이상치(outlier): 사분위에서 특정 범위(+-IQR * 1.5)를 초과하는 경우

### 다른 그래프로 표현하기

- pie graph: 각 범주들이 전체 데이터에서 어느 정도 비율을 차지하는 지

image.png

```python
# pie graph
# plt.pie(비율(ratio), 레이블)
```

```python
pie = food['소분류 카테고리'].value_counts()
pie.values # 비율로 바꾸기
```

```python
# list 값에 대한 계산을 할때는 리스트 컴프리핸션 이용하는 걸 권장
# 순회 가능한 계산이 필요할 때 -> 리스트 컴프리핸션
```

```python
labels = ['한식', '카페', '전문음식', '간이음식', '외국식', '음식점기타']
plt.pie(pie.values, labels=labels, autopct='%.1f%%') # 함승주님 코드
plt.show()
```
